### Removing html tags
___
preparing for data to be much cleaner

```python
def clean_text(line):
    string_name_text = line.replace('<b>','').replace('</b>','').replace('<br/>','\n').split('\n')
    return string_name_text
```

```python
remove_list = ['<span class="code_chat">','\r', '']
dataset_data = []
for item in lists:
    for line in item:
        text_name = clean_text(str(line))
        for value in text_name:
            if value in remove_list:
                pass
            else:
                name_text = re.split(': | - | _',value)
                if len(name_text) > 1:
                    name = str(re.sub('<[^>]*>', '',str(name_text[0])))
                    text = name_text[1]
                    new_text = re.sub('([0-1][0-9]:[0-5][0-9]((\ ){0,1})((AM)|(PM)|(am)|(pm)))|(0[1-9]:[0-5][0-9]((\ ){0,1})((AM)|(PM)|(am)|(pm)))|([1-9]:[0-5][0-9]((\ ){0,1})((AM)|(PM)|(am)|(pm)))|(1[0-2]:[0-5][0-9]((\ ){0,1})((AM)|(PM)|(am)|(pm)))/','',str(text))
                    name = str(re.sub(r'\([^)]*\)', '', name.lower()))
                    text = str(new_text).lower()
                    dataset_data.append([name,text])
                else:
                    print '((((((((((((()))))))))))))'
                    #print re.sub('<.*?>', '',str(name_text))
                    print  str(re.sub('(<span .*>).*(</span>)', '',str(name_text))).lower()
        
```
